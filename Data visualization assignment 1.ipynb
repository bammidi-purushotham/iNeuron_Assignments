```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Reading Data 

dataset = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,4].values
```

```python
# Feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
# Label encoding target values

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Reducing the dimensions to 3D

pca = PCA(n_components=3)
X = pca.fit_transform(X)
```

```python
fig = plt.figure(1,figsize=(4,3))
plt.clf()

ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

# Reorder the labels to have colors matching the cluster results
plt.cla()
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.rainbow_r, edgecolor='k')
plt.show()
```
